    *   `faltantes = ...`: La Serie resultante, que contiene el recuento de valores faltantes por columna, se asigna a la variable `faltantes`.

#### **Celda 5: Código Python (Paso 4)**

````
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()

asimetria = skew(resistencia_limpia)
curtosis_val = kurtosis(resistencia_limpia)

print(f"Asimetría de la Resistencia: {asimetria:.2f}")
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
````

### Explicación del Código para la Forma de la Distribución (Asimetría y Curtosis)

Este bloque de código Python calcula la asimetría (`skewness`) y la curtosis de la variable `Resistencia_MPa` después de manejar los valores faltantes. Estos son indicadores importantes de la forma de la distribución de los datos.

```python
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()

asimetria = skew(resistencia_limpia)
curtosis_val = kurtosis(resistencia_limpia)

print(f"Asimetría de la Resistencia: {asimetria:.2f}")
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
```

**Explicación Línea por Línea:**

1.  `print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")`: Esta línea imprime un encabezado para identificar la sección en la salida del notebook.
2.  `resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()`:
    *   `datos_ingenieria['Resistencia_MPa']`: Selecciona la columna 'Resistencia_MPa' del DataFrame `datos_ingenieria`.
    *   `.dropna()`: Este método elimina las filas que contienen valores faltantes (NaN) específicamente en esta serie. Esto es crucial porque las funciones `skew` y `kurtosis` de `scipy.stats` no manejan valores `NaN` directamente y requieren datos "limpios". El resultado es una nueva Serie llamada `resistencia_limpia`.
3.  `asimetria = skew(resistencia_limpia)`: Calcula el coeficiente de asimetría de la distribución de la `resistencia_limpia`. La asimetría mide la simetría de la distribución. Un valor de asimetría cercano a 0 indica una distribución simétrica. Valores positivos indican una cola más larga a la derecha (datos concentrados a la izquierda), y valores negativos indican una cola más larga a la izquierda (datos concentrados a la derecha).
4.  `curtosis_val = kurtosis(resistencia_limpia)`: Calcula el coeficiente de curtosis de la distribución de la `resistencia_limpia`. La curtosis mide la "puntiagudez" o "aplanamiento" de la distribución, y la presencia de valores atípicos (colas pesadas). Un valor de curtosis (exceso de curtosis) cercano a 0 (para la curtosis de Fisher) indica una distribución similar a la normal (mesocúrtica). Valores positivos (leptocúrtica) indican colas más pesadas y un pico más pronunciado, sugiriendo una mayor probabilidad de valores extremos. Valores negativos (platocúrtica) indican colas más ligeras y un pico más plano.
5.  `print(f"Asimetría de la Resistencia: {asimetria:.2f}")`: Imprime el valor calculado de la asimetría, formateado a dos decimales.
6.  `print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")`: Imprime el valor calculado de la curtosis, formateado a dos decimales.

**Interpretación Ingenieril de la sección:**

La sección incluye una interpretación ingenieril de los resultados de asimetría y curtosis:
*   **Asimetría cercana a 0**: Sugiere que los datos de resistencia están distribuidos de manera relativamente simétrica alrededor de su media. En ingeniería, esto podría indicar que las variaciones en la resistencia del material son balanceadas, sin una tendencia fuerte hacia valores inusualmente altos o bajos.
*   **Curtosis > 0 (colas pesadas)**: Implica una mayor probabilidad de encontrar valores extremos de resistencia (tanto muy altos como muy bajos). En el contexto de materiales, esto es crítico porque una cola pesada en el lado inferior de la resistencia podría indicar una mayor probabilidad de fallos inesperados del material, lo cual es una preocupación de seguridad importante en ingeniería.
