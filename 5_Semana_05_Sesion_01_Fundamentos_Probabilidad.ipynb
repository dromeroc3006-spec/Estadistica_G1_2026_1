## Fundamentos de Probabilidad para Ingeniería

### 1. Contexto en Ingeniería
En las diversas ramas de la ingeniería (civil, agrícola, agroindustrial), la incertidumbre es una constante. La probabilidad se convierte en el lenguaje matemático esencial para cuantificar y modelar esta incertidumbre, permitiéndonos tomar decisiones informadas, evaluar riesgos y diseñar sistemas más robustos frente a la variabilidad de los fenómenos naturales o de proceso.

### 2. Conceptos Fundamentales
*   **Experimento Aleatorio:** Cualquier proceso cuyo resultado no se puede predecir con certeza antes de que ocurra, pero del cual conocemos todos sus posibles resultados. Ejemplos: el resultado de una prueba de resistencia de materiales, la germinación de una semilla, la falla de un componente.
*   **Espacio Muestral ($\Omega$):** El conjunto de todos los posibles resultados de un experimento aleatorio. Por ejemplo, al inspeccionar una viga, $\Omega$ podría ser \{$Aprobado, Fallo\_Menor, Fallo\_Crítico\$\}
*   **Evento ($A$):** Cualquier subconjunto del espacio muestral. Un evento es uno o más resultados específicos de un experimento. Por ejemplo, el evento \{$Fallo\_Critico$\}

### 3. Ecuaciones de Probabilidad

*   **Probabilidad Teórica (Regla de Laplace):**
    Cuando todos los resultados en el espacio muestral son igualmente probables, la probabilidad de un evento $A$ se calcula como:
    $$P(A) = \frac{\text{Número de casos favorables a A}}{\text{Número total de casos posibles}}$$
*   **Probabilidad Empírica (Frecuencia Relativa):**
    Basada en la observación de experimentos repetidos, la probabilidad empírica de un evento $A$ se estima como:
    $$P(A) \approx \frac{\text{Número de veces que ocurre A}}{\text{Número total de repeticiones del experimento}}$$

### 4. Transición Práctica
A continuación, utilizaremos Python para simular experimentos y calcular estas probabilidades empíricas, conectando así la teoría abstracta con la realidad de los datos.

## **Celda 3: Código Python (Espacio Muestral, Eventos y Probabilidad Empírica)**

````
import pandas as pd
import numpy as np

# Simulamos el registro histórico de inspección de 500 vigas de concreto (Ing. Civil)
# o lotes de producción (Ing. Agroindustrial).
# Resultados posibles (Espacio Muestral): "Aprobado", "Fallo_Menor", "Fallo_Critico"

np.random.seed(42)
espacio_muestral = ['Aprobado', 'Fallo_Menor', 'Fallo_Critico']
# Asignamos una probabilidad teórica subyacente (80% éxito, 15% fallo menor, 5% fallo crítico)
inspecciones = np.random.choice(espacio_muestral, size=500, p=[0.80, 0.15, 0.05])

df_calidad = pd.DataFrame({'Estado_Inspeccion': inspecciones})

# ==========================================
# CÁLCULO DE PROBABILIDAD EMPÍRICA
# ==========================================
print("--- FRECUENCIAS ABSOLUTAS ---")
frecuencias = df_calidad['Estado_Inspeccion'].value_counts()
print(frecuencias)

print("\n--- PROBABILIDAD EMPÍRICA (Frecuencia Relativa) ---")
prob_empirica = df_calidad['Estado_Inspeccion'].value_counts(normalize=True)
print(prob_empirica.round(4))

# Interpretación Ingenieril:
# El "Espacio Muestral" son los 3 estados posibles.
# Un "Evento" puede ser que una viga tenga un 'Fallo_Critico'.
# La Probabilidad Empírica de 'Fallo_Critico' calculada con nuestros datos es cercana a 0.05 (5%),
# calculada simplemente dividiendo los casos favorables entre el total de inspecciones.
````

## Probabilidad Condicional e Independencia en Ingeniería

### 1. Introducción Práctica
En ingeniería, no siempre la probabilidad de un evento ocurre de forma aislada. Con frecuencia, la ocurrencia de un evento puede modificar o influir en la probabilidad de otro. Por ejemplo, la probabilidad de que un material falle bajo cierta carga puede depender significativamente del proveedor que lo suministró o del proceso de fabricación utilizado. Entender estas relaciones nos permite afinar nuestros modelos de riesgo y tomar decisiones más precisas.

### 2. Conceptos Clave
*   **Probabilidad Condicional:** Es la probabilidad de que un evento $A$ ocurra, *dado que* otro evento $B$ ya ha ocurrido. Nos permite actualizar nuestra incertidumbre sobre $A$ con nueva información sobre $B$. En la práctica, esto podría ser la probabilidad de un fallo crítico, dado que el material proviene de un Proveedor_B.
*   **Independencia Estadística:** Dos eventos $A$ y $B$ son estadísticamente independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. En un contexto de control de calidad, si la tasa de fallos de un producto es independiente del proveedor, significa que la elección del proveedor no influye en la calidad del producto, lo que es un hallazgo importante.

### 3. Ecuaciones en LaTeX
*   **Probabilidad Condicional:** La probabilidad de $A$ dado $B$ se define como:
    $$P(A|B) = \frac{P(A \cap B)}{P(B)}$$
    Donde $P(A \cap B)$ es la probabilidad de que ambos eventos $A$ y $B$ ocurran, y $P(B)$ es la probabilidad de que el evento $B$ ocurra ($P(B) > 0$).
*   **Condición Matemática de Independencia:** Los eventos $A$ y $B$ son independientes si se cumple cualquiera de las siguientes condiciones:
    $$P(A|B) = P(A)$$
    O, equivalentemente:
    $$P(A \cap B) = P(A) \cdot P(B)$$

### 4. Transición al Código (Tablas de Contingencia)
A continuación, utilizaremos la biblioteca Pandas de Python para construir una tabla de contingencia cruzada, lo que nos permitirá calcular y analizar estas probabilidades condicionales a partir de nuestros datos simulados de inspecciones y proveedores.

## **Celda 4: Código Python (Probabilidad Condicional e Independencia)**

````
# Ahora, supongamos que estas inspecciones provienen de dos proveedores distintos
proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])
df_calidad['Proveedor'] = proveedores

# Tabla de contingencia cruzada
tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)

print("--- TABLA DE CONTINGENCIA ---")
print(tabla_cruzada)

# Cálculo de Probabilidad Condicional
# P(Fallo_Critico | Proveedor_B) = Casos de Fallo Crítico en Prov_B / Total de Prov_B
total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']
fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']

prob_condicional = fallos_prov_b / total_prov_b
print(f"\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}")

# Interpretación: Si la probabilidad condicional es igual a la probabilidad marginal (general),
# decimos que los eventos son INDEPENDIENTES. En la práctica ingenieril, si un proveedor
# tiene una tasa de fallos mucho mayor, la calidad "depende" del proveedor.
````
